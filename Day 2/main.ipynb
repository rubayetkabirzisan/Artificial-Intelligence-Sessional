BFS : Breadth First Search

```python
graph ={
    '5':['3','7'],
    '3':['2','4'],
    '7':['8'],
    '2':[],
    '4':['8'],
    '8':[],
}
queue=[]
visited=[]
def bfs(visited, graph,node):
  queue.append(node)
  visited.append(node)
  while queue:
    m=queue.pop(0)
    print(m, end=" ")
    for neighbour in graph[m]:
      if neighbour not in visited:
        visited.append(neighbour)
        queue.append(neighbour)

print("Following is the Breadth-First Search :")
bfs(visited, graph,'5')
```

```python
from collections import deque
graph ={
    'S':['A','H'],
    'A':['B','C'],
    'B':['D','E'],
    'C':['F','G'],
    'H':['I'],
    'I':['J'],
    'D':[],
    'E':[],
    'F':[],
    'G':[],
    'J':[],
}
queue=deque()
visited=[]
def bfs(visited, graph,node):
  queue.append(node)
  visited.append(node)
  while queue:
    m=queue.popleft()
    print(m, end=" ")
    for neighbour in graph[m]:
      if neighbour not in visited:
        visited.append(neighbour)
        queue.append(neighbour)

print("Following is the Breadth-First Search :")
bfs(visited, graph,'S')
```
