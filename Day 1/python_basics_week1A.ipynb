# **Variables & Data Types**
###Variables:
In python, variables are like containers for storing data values. You don't need to declare variable's type. The type is inferred from the value you assign.

###Common Data Types:
*   **Integers (`int`)**: Whole numbers (e.g., `5`, `-100`).
*   **Floating-point numbers (`float`)**: Numbers with a decimal point (e.g., `3.14`, `-0.5`).
*   **Strings (`str`)**: Sequences of characters, enclosed in single or double quotes or triple quotes (e.g., `'hello'`, `"Python"`, `"""Morning"""`).
*   **Booleans (`bool`)**: Represents truth values, either `True` or `False`.


```python
# Integers
age = 30
num_students = 25
print(f"Age: {age}, Type: {type(age)}")
print("Output: ", age, "Type: ", type(age))

# Floats
price = 19.99
pi = 3.14159
print(f"Price: {price}, Type: {type(price)}")

# Strings
name = "Alice"
greeting = 'Hello, world!'
print(f"Name: {name}, Type: {type(name)}")

# Booleans
is_active = True
has_license = False
print(f"Is Active: {is_active}, Type: {type(is_active)}")
```

#**Data Structures**
Python has built-in data structures that organize and store collections of data.

###List:
* Ordered and mutable collection of items
* Items are enclosed within square brackets (`[]`)
* Items can be of different data types.


```python
myList = [1, 'apple', 20.39] #--> multiple types of data can be collected in a List
#myList = list(map(int, input().split()))
print("Original List: ", myList)

#Accessing Elements
print("First Element: ", myList[0])
print("Second Element: ", myList[1])

# Modifying elements
myList[0] = 100
print("Modified list: ", myList)

# Adding elements
myList.append("orange")  #Adds value at the end
print("List after append: ", myList)

myList.insert(1, "banana")  #Adds value at given position .insert(position, value)
print("List after insert: ", myList)

# Removing elements
myList.remove(20.39)
print("List after remove: ", myList)

myList.pop(1)
print("List after pop: ", myList)

# Length of the list
print("Length of the list: ", len(myList))
```

###Tuples
*   Ordered and immutable collection of items.
*   Items are enclosed in parentheses `()`.

```python
my_tuple = (10, 20, 30, "grape")
print(f"Original tuple: {my_tuple}")

# Accessing elements
print(f"First element: {my_tuple[0]}")

# Modifying a tuple will raise an error
#my_tuple[0] = 100
```

### Dictionaries

*   Unordered, changeable, and indexed collection of items.
*   Store data in `key: value` pairs.
*   Items are enclosed in curly braces `{}`.

```python
myDictionary = {}
```

```python
myDict = {"name": "Bob", "age": 25, "city": "New York"}
print("Original dictionary: ", myDict)

# Accessing values by key
print("Name: ", myDict['name'])
print("City: ", myDict.get('city'))

# Modifying values
myDict['age'] = 26
print("Modified dictionary: ", myDict)

# Adding new key-value pairs
myDict['occupation'] = "Engineer"
print("Dictionary after adding: ", myDict)

# Removing key-value pairs
del myDict['city']
print("Dictionary after removing: ", myDict)
```

### Sets

*   Unordered and unindexed collection of unique items.
*   Items are enclosed in curly braces `{}` (or `set()` for an empty set).
*   Automatically removes duplicate elements.

```python
mySet = {1, 2, 3, 2, 1, "hello", "world"}
print("Original set (duplicates removed): ", mySet)

# Adding elements
mySet.add(4)
print("Set after adding: ", mySet)

# Removing elements
mySet.remove(3)
print("Set after removing: ", mySet)

# Set operations
setA = {1, 2, 3, 4}
setB = {3, 4, 5, 6}
print("Union: ", setA.union(setB))          # Elements in either set
print("Intersection: ", setA.intersection(setB)) # Elements in both sets
print("Difference: ", setA.difference(setB))    # Elements in set_a but not in set_b
```

# **Control Flow**

Control flow statements determine the order in which the code is executed. They allow your program to make decisions and repeat actions.

### `if`, `elif`, `else` Statements

Used for conditional execution. The code block under an `if`, `elif`, or `else` statement is executed only if its condition is `True`.

```python
score = 66

if score >= 90:
    print("Grade: A")
elif score >= 80:
    print("Grade: B")
elif score >= 70:
    print("Grade: C")
else:
    print("Grade: F")
```

### `for` Loops

Used for iterating over a sequence (like a list, tuple, string, or range) or other iterable objects.

```python
# Iterating over a list
fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    print(f"I like {fruit}")

# Iterating using range()
# range(5) generates numbers from 0 up to (but not including) 5
print("\nNumbers from 0 to 4:")
for num in range(5):
    print(num)

# Iterating with index and value using enumerate()
print("\nFruits with index:")
for index, fruit in enumerate(fruits):
    print(f"{index}: {fruit}")
```

### `while` Loops

Repeats a block of code as long as a certain condition is true.

```python
count = 0
while count < 5:
    print(f"Count is: {count}")
    count += 1 # This is equivalent to count = count + 1

print("Loop finished.")
```

## 7. Functions

Functions are blocks of organized, reusable code that perform a single, related action. They help break down your program into smaller, manageable parts, making it more organized and efficient.

### Defining a Function

Functions are defined using the `def` keyword, followed by the function name, parentheses `()`, and a colon `:`.

```python
# A simple function that prints a greeting
def greet():
    print("Hello there!")

# A function that takes an argument (parameter)
def greet_name(name):
    print(f"Hello, {name}!")

# A function that returns a value
def add_numbers(a, b):
    return a + b
```

### Calling a Function

To execute the code inside a function, you call it by its name followed by parentheses `()`.

```python
# Calling the greet function
greet()

# Calling the greet_name function with an argument
greet_name("Alice")
greet_name("Bob")

# Calling the add_numbers function and storing its return value
result = add_numbers(5, 7)
print("The sum is: ", result)

print("Another sum: ", add_numbers(10, 20))
```
