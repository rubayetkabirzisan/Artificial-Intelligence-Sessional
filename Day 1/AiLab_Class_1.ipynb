```python
marks = eval(input("Enter the Grade: "))

if marks <= 10:
  print(f"Grade: Fail")
elif marks <= 30:
  print(f"Grade: D")
elif marks <= 50:
  print(f"Grade: C")
elif marks <= 70:
  print(f"Grade: B")
else:
  print(f"Grade: A")
```

```python
# for count in range (initial, end, steps)
# steps means how much to increment or decrement
# if not specified then will increase 1 in each step
# the end is not inclusive

n = eval(input("Enter the Number: "))
sum = 0

for count in range (1, n+1):
  sum = sum + count

print(f"The sum is {sum}")
```

```python
# to take multiple types of datas in a single array --> List

# mylist = list(map(int, input().split()))

mylist = [1, "Coding", 98.89]

for i in range(len(mylist)):
  print(f"{mylist[i]}")

# Modifying elements
mylist[1] = "Code-Modified"

print("\nAfter Modification: ")

for i in range(len(mylist)):
  print(f"{mylist[i]}")

# Appending elements --> Adding at the last
mylist.append("Appended")

print("\nAfter Appending: ")

for i in range(len(mylist)):
  print(f"{mylist[i]}")

# Removing elements
print("\nAfter Removing by value: ")

mylist.remove(98.89)

for i in range(len(mylist)):
  print(f"{mylist[i]}")

print("\nAfter Removing by position: ")

mylist.pop(len(mylist)-1)

for i in range(len(mylist)):
  print(f"{mylist[i]}")
```

```python
# Making a function - 4 Ways

def summingNumber(): # No Parameter , No returns
  numberOne = eval(input("Enter the First Number: "))
  numberTwo = eval(input("Enter the Second Number: "))

  sum = numberOne + numberTwo

  print(f"The addition result is {sum}")

# For writing main function
def main():
  summingNumber()

if __name__ == "__main__":
  main()

# If I don't want a main function I can directly write
# summingNumber()
```

```python
def convertCelciusToFarenhite(): # Without main function
  celcius = eval(input("Enter the Celcius: "))

  farenhite = (celcius * 9/5) + 32

  print(f"The Farenhite is {farenhite}")

convertCelciusToFarenhite()
```

```python
def convertCelciusToFarenhite(): # With main function
  celcius = eval(input("Enter the Celcius: "))

  farenhite = (celcius * 9/5) + 32

  print(f"The Farenhite is {farenhite}")

def main():
  convertCelciusToFarenhite()

if __name__ == "__main__":
  main()
```

```python
def summingNumber(numberOne, numberTwo): # With Parameter , No returns
  sum = numberOne + numberTwo
  print(f"The addition result is {sum}")

numberOne = eval(input("Enter the First Number: "))
numberTwo = eval(input("Enter the Second Number: "))

summingNumber(numberOne, numberTwo)
```

```python
def summingNumber(numberOne, numberTwo): # With Parameter and Return
  sum = numberOne + numberTwo
  return sum

numberOne = eval(input("Enter the First Number: "))
numberTwo = eval(input("Enter the Second Number: "))

sum = summingNumber(numberOne, numberTwo)

print(f"The addition result is {sum}")
```
