```python
def summingNumber(): # No parameter no returns
  number1 = eval(input("Enter a number"))
  number2 = eval(input("Enter a number"))
  sum = number1 + number2
  print("Summation ", sum)
```

```python
summingNumber()
```

```python
def summingNumber(number1, number2): #Only parameters no returns
  sum = number1 + number2
  print("Summation ", sum)
```

```python
number1 = eval(input("Enter a number"))
number2 = eval(input("Enter a number"))
summingNumber(number1,number2)
```

```python
def summingNumber(number1, number2): # With parameters with returns
  sum = number1 + number2
  return sum
```

```python
num1 = eval(input("Enter a number"))
num2 = eval(input("Enter a number"))
print("Summation= ", summingNumber(num1,num2))
```
