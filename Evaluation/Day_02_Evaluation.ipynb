```python
import random

# Create empty board
board = [["#" for _ in range(3)] for _ in range(3)]

def show_board():
    print("\nBoard:")
    for row in board:
        print(" ", *row)
    print()

def winner(p):
    # Check rows
    for r in range(3):
        if board[r][0] == board[r][1] == board[r][2] == p:
            return True
    # Check columns
    for c in range(3):
        if board[0][c] == board[1][c] == board[2][c] == p:
            return True
    # Check diagonals
    if board[0][0] == board[1][1] == board[2][2] == p:
        return True
    if board[0][2] == board[1][1] == board[2][0] == p:
        return True
    return False

def empty_cells():
    return [(r, c) for r in range(3) for c in range(3) if board[r][c] == "#"]

print("Welcome! You = X, Computer = O")
show_board()

while True:

    # --------------------- PLAYER MOVE ---------------------
    while True:
        try:
            r, c = map(int, input("Your move (row col): ").split())
            if (r in range(3) and c in range(3) and board[r][c] == "#"):
                board[r][c] = "X"
                break
            print("Invalid move! Try again.")
        except:
            print("Enter two numbers like: 1 2")

    show_board()

    if winner("X"):
        print("You win!")
        break

    if not empty_cells():
        print("Draw!")
        break

    # --------------------- COMPUTER MOVE ---------------------
    print("Computer thinking...")
    r, c = random.choice(empty_cells())
    board[r][c] = "O"

    show_board()

    if winner("O"):
        print("Computer wins!")
        break

    if not empty_cells():
        print("Draw!")
        break
```
